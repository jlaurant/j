```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
```

## Import the input and target output
Both input and output are normalized

```python
# import data
path = 'MBE2019/'
file1 = 'norminputs2.csv'
file2 = 'normoutput2.csv'
X = pd.read_csv(path+file1)
y = pd.read_csv(path+file2)
```

```python
X.head()
```

```python
y.head()
```

## Create training and test set - using holdout method
Holdout method: simply leave out a subset of the data for test set.
Here, we keep 70% of the dataset as training set and 30% as test set.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

## Build the neural network
#### with two hidden layers and 4 neurons each layer
#### using relu as the activation function
#### using mean squared error (mse) as loss function

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ncol=X_train.shape[1]
```

```python
# construct the structure of the neural network
act='relu'
num=4
model = Sequential()
model.add(Dense(num,activation=act, input_shape = (ncol,)))
model.add(Dense(num, activation=act))
model.add(Dense(1, activation=act)) 
model.compile(loss='mse', optimizer='adam')
model.summary()
```

```python
# fit the model and get test results
history = model.fit(X_train,
                    y_train,
                    epochs=300,
                    batch_size=1,
                    validation_data=(X_test, y_test))
```

## Results discussion

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Test loss',markersize=1)
plt.title('Training and test loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

#### The loss (mean squared error) gradualy reduced with each iteration.
#### But is the neural network overfitting? Let's have a close look after 100 epochs.
Overfitting happens when the model follows the training data too closely and performs poorly on test data.

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Test loss',markersize=1)
plt.title('Training and test loss')
plt.xlabel('Epochs')
plt.xlim(100,200)
plt.ylim(0,1000)
plt.ylabel('Loss')
plt.legend()
plt.show()
```

#### We can observe a gap between the training loss and test loss. What does it mean?
It means the neural network suffers from high variance. Getting more data will help with this program.
